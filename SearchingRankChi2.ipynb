적합도 검정~

```python
# 적합도검정
# ~이라고 주장하는 그룹의비율 vs 관측데이터비율 검정
```

```python
import pandas as pd
from scipy.stats import chisquare

file_path = "SearchingRank.csv"
data = pd.read_csv(file_path, encoding="cp949")

# 특정 열에서 카테고리별 빈도 계산 (예: "카테고리" 열)
observed_counts = data["구분"].value_counts()

# 관찰값 기대값 정의
observed = observed_counts.values  # 관찰값
expected = [sum(observed) / len(observed)] * len(observed)  # 균등 분포 가정

# 카이제곱 적합도 검정 수행
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# 결과 출력
print("Chi-squared statistic:", chi2_stat)
print("p-value:", p_value)

# 관찰값 및 기대값 확인
print("\nObserved counts:")
print(observed_counts)
print("\nExpected counts (equal distribution):")
print(expected)
```

```python
from scipy.stats import chi2

# 자유도와 유의 수준 설정
dof = 2  # 자유도
alpha = 0.05  # 유의 수준

# 카이제곱 기각값 (임계값) 계산
critical_value = chi2.ppf(1 - alpha, dof)

# 결과 출력
print(f"기각값 (카이제곱 임계값) at alpha={alpha} and df={dof}: {critical_value}")
```

```python
# 카이제곱 검정통계량 Chi-squared statistic: 110.6292
# 신뢰수준 5% -> 유의수준 α:0.05
# p-value: 3.4052*10^-16, dof:3-1=2

# 기각값 C.V.:27.5871 < 검정통계량 110.6292, H₀기각

# 주장과 다르게 외지인과 현지인의 비율이 균등하지 않다(구분의전체=현지인+외지인으로가정)
# H₀:모집단이 균등분포를 따른다(기각)
# Η₁:모집단이 균등분포를 따르지 않는다(선택)
```

독립성 검정~

```python
# 이변수분할표, 변수: 카테고리(X), 검색순위(Y)
# n>30 이므로 대표본 가정을 만족한다고 보고

df_cross = pd.crosstab(df['카테고리'], df['순위'])
df_cross
```

```python
from scipy.stats import chi2

# 자유도 (df) 계산
r = 17  # 행의 수 (예: 변수1의 카테고리 수)
c = 25  # 열의 수 (예: 변수2의 카테고리 수)
df = (r - 1) * (c - 1)

# 유의수준 설정
alpha = 0.05  # 일반적으로 0.05 사용

# 카이제곱 기각값 (임계값) 계산
critical_value = chi2.ppf(1 - alpha, df)

# 결과 출력
print(f"자유도(df)={df}, 유의수준(alpha)={alpha}에 대한 카이제곱 기각값: {critical_value}")
```

```python
# 카이스퀘어 분포를 사용한 이원카이제곱검정, 독립성검정
# 신뢰도 95%에서 가설검증=유의수준 α=0.05(100%-95%)

# statistic=665.3495, p-value=2.1222*10^-17, dof=384
# p-value < 0.05, 1종 오류확률 5%미만으로 귀무가설 기각

# (카테고리)행과 (검색순위)열이 서로 영향을 준다
# H₀귀무가설: 카테고리와 검색순위는 관련성이 없다(기각)
# Η₁대립가설: 카테고리와 검색순위는 관련성이 있다(선택)

input= np.array(df_cross)
stats.chi2_contingency(observed=input) #,correction=True,lambda_=None
```
